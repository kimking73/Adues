```python
import nltk
nltk.download('all')
```

```python
import json
import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
nltk.download('punkt')
```

```python
intents = {
    "intents": [
        {
            "tag": "인사",
            "patterns": ["안녕", "안녕하세요", "하이", "반가워", "안뇽"],
            "responses": ["안녕하세요! 반가워요.", "안녕~ 좋은 하루에요!", "하이~ 무엇을 도와줄까요?"]
        },
        {
            "tag": "날씨",
            "patterns": ["오늘 날씨 어때?", "비 와?", "날씨 알려줘", "기온 어때?", "우산 필요해?"],
            "responses": ["날씨는 앱으로 확인해보세요! 오늘도 좋은 하루 되세요.", "날씨는 잘 모르지만, 기분은 맑았으면 좋겠어요!", "비가 올 수도 있으니 우산 챙기세요~"]
        },
        {
            "tag": "위로",
            "patterns": ["기분이 안 좋아", "속상해", "힘들어", "지쳤어", "우울해"],
            "responses": ["힘들었겠어요. 잠깐 쉬어가는 것도 좋아요.", "괜찮아요, 다 잘 될 거예요.", "오늘도 잘 버텨줘서 고마워요."]
        }
    ]
}
```

```python
texts = []
labels = []

for intent in intents["intents"]:
    for pattern in intent["patterns"]:
        texts.append(pattern)
        labels.append(intent["tag"])

le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

model = Pipeline([
    ('vectorizer', CountVectorizer(tokenizer=nltk.word_tokenize)),
    ('classifier', MultinomialNB())
])

model.fit(texts, labels_encoded)
```

```python
def chatbot_response(user_input):
    pred = model.predict([user_input])[0]
    tag = le.inverse_transform([pred])[0]

    for intent in intents["intents"]:
        if intent["tag"] == tag:
            return random.choice(intent["responses"])
```

```python
print("👋 안녕하세요! 덕배 챗봇이에요. (종료하려면 '그만' 입력)")

while True:
    user_input = input("너: ")
    if user_input.lower() == "그만":
        print("덕배: 다음에 또 만나요!")
        break
    response = chatbot_response(user_input)
    print("덕배:", response)
```
