```python
import json
import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
nltk.download('punkt')
```

```python
intents = {
    "intents": [
        {
            "tag": "Ïù∏ÏÇ¨",
            "patterns": ["ÏïàÎÖï", "ÏïàÎÖïÌïòÏÑ∏Ïöî", "ÌïòÏù¥", "Î∞òÍ∞ÄÏõå", "ÏïàÎáΩ"],
            "responses": ["ÏïàÎÖïÌïòÏÑ∏Ïöî! Î∞òÍ∞ÄÏõåÏöî.", "ÏïàÎÖï~ Ï¢ãÏùÄ ÌïòÎ£®ÏóêÏöî!", "ÌïòÏù¥~ Î¨¥ÏóáÏùÑ ÎèÑÏôÄÏ§ÑÍπåÏöî?"]
        },
        {
            "tag": "ÎÇ†Ïî®",
            "patterns": ["Ïò§Îäò ÎÇ†Ïî® Ïñ¥Îïå?", "ÎπÑ ÏôÄ?", "ÎÇ†Ïî® ÏïåÎ†§Ï§ò", "Í∏∞Ïò® Ïñ¥Îïå?", "Ïö∞ÏÇ∞ ÌïÑÏöîÌï¥?"],
            "responses": ["ÎÇ†Ïî®Îäî Ïï±ÏúºÎ°ú ÌôïÏù∏Ìï¥Î≥¥ÏÑ∏Ïöî! Ïò§ÎäòÎèÑ Ï¢ãÏùÄ ÌïòÎ£® ÎêòÏÑ∏Ïöî.", "ÎÇ†Ïî®Îäî Ïûò Î™®Î•¥ÏßÄÎßå, Í∏∞Î∂ÑÏùÄ ÎßëÏïòÏúºÎ©¥ Ï¢ãÍ≤†Ïñ¥Ïöî!", "ÎπÑÍ∞Ä Ïò¨ ÏàòÎèÑ ÏûàÏúºÎãà Ïö∞ÏÇ∞ Ï±ôÍ∏∞ÏÑ∏Ïöî~"]
        },
        {
            "tag": "ÏúÑÎ°ú",
            "patterns": ["Í∏∞Î∂ÑÏù¥ Ïïà Ï¢ãÏïÑ", "ÏÜçÏÉÅÌï¥", "ÌûòÎì§Ïñ¥", "ÏßÄÏ≥§Ïñ¥", "Ïö∞Ïö∏Ìï¥"],
            "responses": ["ÌûòÎì§ÏóàÍ≤†Ïñ¥Ïöî. Ïû†Íπê Ïâ¨Ïñ¥Í∞ÄÎäî Í≤ÉÎèÑ Ï¢ãÏïÑÏöî.", "Í¥úÏ∞ÆÏïÑÏöî, Îã§ Ïûò Îê† Í±∞ÏòàÏöî.", "Ïò§ÎäòÎèÑ Ïûò Î≤ÑÌÖ®Ï§òÏÑú Í≥†ÎßàÏõåÏöî."]
        }
    ]
}
```

```python
texts = []
labels = []

for intent in intents["intents"]:
    for pattern in intent["patterns"]:
        texts.append(pattern)
        labels.append(intent["tag"])

le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

model = Pipeline([
    ('vectorizer', CountVectorizer(tokenizer=nltk.word_tokenize)),
    ('classifier', MultinomialNB())
])

model.fit(texts, labels_encoded)
```

```python
def chatbot_response(user_input):
    pred = model.predict([user_input])[0]
    tag = le.inverse_transform([pred])[0]

    for intent in intents["intents"]:
        if intent["tag"] == tag:
            return random.choice(intent["responses"])
```
